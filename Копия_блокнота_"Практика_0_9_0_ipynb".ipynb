# Задание 3


Цель - создать менеджера команды Pokémon, который позволит пользователям:

- Добавлять покемонов в свою команду. (если такого покемона еще нет в команде)
- Удалять покемонов из их команды.
- Просматривать подробную информацию обо всех покемонах в команде.
- Находить покемона по имени.
- Устраивать тренировочный бой между двумя покемонами

Для данной задачи используйте: https://pokeapi.co/


````python
import requests


def get_pokemon(name):
	try:
		url = f"https://pokeapi.co/api/v2/pokemon/{name}"
		response = requests.get(url)
		if response.status_code == 200:
			data = response.json()
			return {
				'name': data['name'],
				'types': [t['type']['name'] for t in data['types']],
				'hp': data['stats'][0]['base_stat'],
				'attack': data['stats'][1]['base_stat'],
				'defense': data['stats'][2]['base_stat']
			}
		return None
	except:
		return None


def main():
	team = []

	while True:
		print("1. Добавить покемона")
		print("2. Удалить покемона")
		print("3. Показать команду")
		print("4. Устроить бой")
		print("5. Выход")

		choice = input("Выберите действие ")


		if choice == '1':

			name = input("Введите имя покемона ")
			pokemon = get_pokemon(name)

			if not pokemon:
				print("Покемон не найден")
				continue

			if any(p['name'] == pokemon['name'] for p in team):
				print("Такой покемон уже есть в команде")
				continue

			team.append(pokemon)
			print(f"{pokemon['name']} добавлен в команду")

		elif choice == '2':
			if not team:
				print("Команда пуста")
				continue

			name = input("Введите имя покемона: ")
			before_len = len(team)
			team = [p for p in team if p['name'] != name]

			if len(team) == before_len:
				print("Такой покемон не найден в команде")
			else:
				print(f"{name} удален из команды")

		elif choice == '3':
			if not team:
				print("Команда пуста")
				continue

			for i, pokemon in enumerate(team, 1):
				print(f"{i}. {pokemon['name']}")
				print(f"Типы: {', '.join(pokemon['types'])}")
				print(f"HP: {pokemon['hp']}")
				print(f"Атака: {pokemon['attack']}")
				print(f"Защита: {pokemon['defense']}")

		elif choice == '4':
			if len(team) < 2:
				print("Нужно минимум 2 покемона для боя")
				continue

			print("Ваша команда:")
			for i, pokemon in enumerate(team, 1):
				print(f"{i}. {pokemon['name']}")

			try:
				first_ID = int(input("Выберите номер первого покемона ")) - 1
				second_ID = int(input("Выберите номер второго покемона ")) - 1

				pokemon_1 = team[first_ID]
				pokemon_2 = team[second_ID]

				power_1 = pokemon_1['hp'] + pokemon_1['attack'] + pokemon_1['defense']
				power_2 = pokemon_2['hp'] + pokemon_2['attack'] + pokemon_2['defense']

				print(f"{pokemon_1['name']} (сила: {power_1}) vs \
         {pokemon_2['name']} (сила: {power_2})")

				if power_1 > power_2:
					print(f"Победитель {pokemon_1['name']}")
				elif power_2 > power_1:
					print(f"Победитель {pokemon_2['name']}")
				else:
					print("Ничья")

			except:
				print("Ошибка,проверьте номера покемонов")

		elif choice == '5':
			break

		else:
			print("Неверный выбор")


if __name__ == "__main__":
	main()
````
